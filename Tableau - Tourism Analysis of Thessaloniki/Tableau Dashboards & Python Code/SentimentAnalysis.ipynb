```python
def extract_keyword(review):
    r.extract_keywords_from_text(str(review))
    phrases_with_scores = r.get_ranked_phrases_with_scores()
    if phrases_with_scores:
        no1_phrase = phrases_with_scores[0][1]
        words = nltk.word_tokenize(no1_phrase)
        return words[0]
    else:
        return ''
#Αυτή η συνάρτηση εξάγει μια λέξη-κλειδί που περιγράφει καλύτερα το συναίσθημα της κριτικής
```

```python
def sentiment(review):
    sentiment_score = sentiment_analyzer.polarity_scores(str(review))['compound']
    sentiment_score = sentiment_analyzer.polarity_scores(str(review))['compound']
    if sentiment_score>0.2:
        sentiment='Positive'
    elif sentiment_score <-0.2:
        sentiment='Negative'
    else:
        sentiment='Neutral'
    return sentiment
#Αυτή η συνάρτηση ταξινομεί το συναίσθημα του κειμένου σε "Θετικό", "Ουδέτερο" και "Αρνητικό" με βάση τις βαθμολογίες 
#μεταξύ -1 προς 1.
```

```python
df['Sentiment'] = df['Title'].apply(sentiment)
```

```python
df['Keyword'] = df['Title'].apply(extract_keyword)
```
